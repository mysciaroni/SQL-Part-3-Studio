### **Write a query that returns tag\_id, the number of times each tag\_id is used and the tag\_name.**  

### Use the GROUP BY and HAVING clause to narrow your focus.

Try multiple keywords, ie woman vs female (female ended up giving better results)

```sql
-- Solution
USE BooksDB
SELECT book_tags.tag_id AS 'TAG ID' ,tags.tag_name AS 'TAG NAME',
SUM(book_tags.count) AS 'TAG COUNT'
FROM book_tags
INNER JOIN tags
ON book_tags.tag_id=tags.tag_id
GROUP by book_tags.tag_id, tags.tag_name 
HAVING tag_name LIKE '%WOMAN%'
ORDER BY SUM(book_tags.count) DESC;

--Attempting to see difference between Inner Join and Full Join

USE BooksDB
SELECT book_tags.tag_id AS 'TAG ID' ,tags.tag_name AS 'TAG NAME',
SUM(book_tags.count) AS 'TAG COUNT'
FROM book_tags
FULL OUTER JOIN tags
ON book_tags.tag_id=tags.tag_id
GROUP by book_tags.tag_id, tags.tag_name 
HAVING tag_name LIKE '%WOMAN%'
ORDER BY SUM(book_tags.count) DESC;
```

### **Write a query to return authors, titles, ratings and** **selected tag(s)\_id.**

```sql
-- Solution
USE BooksDB
SELECT book_tags.tag_id AS 'TAG ID', books.authors, books.title, books.average_rating
FROM book_tags
INNER JOIN books
ON book_tags.goodreads_book_id=books.book_id
WHERE book_tags.tag_id = 11643
ORDER BY books.authors, books.title, books.average_rating DESC
-- GROUP by book_tags.tag_id, books.title
-- HAVING tag_id LIKE '%WOMAN%'
-- ORDER BY SUM(book_tags.count) DESC;

-- USE BooksDB
-- SELECT book_tags.tag_id AS 'TAG ID', books.authors, count(books.authors), books.title, books.average_rating
-- FROM book_tags
-- INNER JOIN books
-- ON book_tags.goodreads_book_id=books.book_id
-- WHERE book_tags.tag_id = 11643
-- GROUP BY books.authors, books.title, books.average_rating
-- ORDER BY books.authors, books.title, books.average_rating DESC
```

```sql
-- Event 1 Query - I want to just see in general how many books have a tag_name that has to do with LGBTQ
USE BooksDB
SELECT book_tags.tag_id AS 'TAG ID' ,tags.tag_name AS 'TAG NAME',
SUM(book_tags.count) AS 'TAG COUNT'
FROM book_tags
INNER JOIN tags
ON book_tags.tag_id=tags.tag_id
GROUP by book_tags.tag_id, tags.tag_name 
HAVING tag_name LIKE '%LGBTQ%'
ORDER BY SUM(book_tags.count) DESC;
```

```sql
USE BooksDB
SELECT tags.tag_name, tags.tag_id, books.book_id, books.title, books.authors, books.average_rating
FROM tags, books
WHERE tag_name = 'ya-lgbtq'
GROUP BY tags.tag_name, tags.tag_id, books.book_id, books.title, books.authors, books.average_rating
ORDER BY books.average_rating DESC;
```
